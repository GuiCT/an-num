# Exercício 1.

A propriedade da simetria dos pesos em métodos ed Newton-Cotes com pontos igualmente espaçados é resultado de uma outra simetria, presente nos interpoladores de Lagrange associados a tais pontos.

Por exemplo, dado um conjunto de pontos igualmente espaçados $(x_0, x_1, \ldots, x_n)$ no intervalo $[a, b]$, onde $x_i = a + i \cdot h$ e $h = \displaystyle\frac{b - a}{n}$, os polinômios de Lagrange $l_i(x)$ associados a esses pontos possuem a seguinte propriedade de simetria:

$$l_i(x) = l_{n-i}(a + b - x)$$

Realizando a substituição

$$ u = a + b - x \Rightarrow \frac{du}{dx} = -1 \Rightarrow du = -dx $$

Podemos a partir do valor de $w_{n-k}$, chegar no mesmo valor de $w_k$:

$$ w_{n-k} = \int_a^b l_{n - k} (x)\ dx = - \int_b^a l_{n-k} (a + b - x)\ du = - \int_b^a l_k (x)\ du = \int_a^b l_k (x)\ dx = w_k\ \blacksquare $$

# Exercício 2.

Definindo $X = (a + b)/2$ e realizando a substituição $ u = x - X $, com $du = dx$, transforma-se a integral para o intervalo dado por:

$$ \Rightarrow a - \frac{a + b}{2} = \frac{a - b}{2} = -L $$
e
$$ \Rightarrow b - \frac{a + b}{2} = \frac{b - a}{2} = L $$

dessa maneira, a integral é transformada para esse novo intervalo:
$$ \int_a^b (x - X)^{n + 1}\ dx \Rightarrow \int_{-L}^L u^{n + 1}\ du $$

Como $n$ é par, logo $n+1$ é ímpar. Dessa maneira, $g(u) = u^{n + 1}$ é uma função ímpar, e a integral de uma função ímpar em um intervalo simétrico em relação à origem é sempre zero:

$$ \int_{-L}^L u^{n + 1}\ du = 0 $$

O cálculo a partir de Newton-Cotes é dada por uma expressão do tipo

$$ Q = \sum_{k=0}^n w_k f(x_k) $$

Como definimos $X$ de forma que ele representa o ponto médio do intervalo, existe a simetria:

$$ x_k - X = -(x_{n - k} - X) \Rightarrow x_k - X = X - x_{n - k} $$

Logo, a aplicação $f(x_{n - k}) = (x_{n - k} - X)^{n + 1} = (-(x_k - X))^{n + 1}$ e a aplicação $f(x_k) = (x_k - X)^{n + 1}$ estão relacionadas.

Como $f$ é ímpar tal que $f(-x) = -f(x)$, obtemos a seguinte equivalência:

$$ f(x_{n - k}) = -(x_k - X)^{n + 1} = -f(x_k) $$

Dada a fórmula de Newton-Cotes:

$$ Q = \sum_{k = 0}^n w_k f(x_k) $$

Dado o termo central $x_{n/2}$ e sabendo da simetria dos pesos, é possível realizar a combinação:

$$ Q = (w_0 f(x_0) + w_n f(x_n)) + (w_1 f(x_1) + w_{n-1} f(x_1)) + \dots + w_{n/2} f(x_{n/2}) $$

Pela simetria de pesos e simetria dos termos simétricos:

$$ w_k f(x_k) + w_{n - k} f(x_n - k) = w_k f(x_k) + w_k (-f(x_k)) = w_k f(x_k) - w_k f(x_k) = 0 $$

E como $x_{n/2} = X$, logo $f(x_{n/2}) = (x_{n/2} - X)^{n + 1} = 0^{n + 1} = 0 $

Dessa maneira, confirmamos que a soma total da quadratura é zero. Como $I=Q=0$, a fórmula de Newton-Cotes é exata para polinômios de grau $n + 1$ quando $n$ é par $\blacksquare$.

# Exercício 3.

```python
x3 = np.arange(0, 22, 2, dtype=np.float64)
y3 = np.array([0, 1.8, 2, 4, 4, 6, 4, 3.6, 3.4, 2.8, 0])
h = x3[1] - x3[0]

# Utilizando fórmula de Simpson Composta
sum_value = 0.0
k = 0
while k + 2 < len(y3):
    to_sum = (h / 3) * (y3[k] + 4 * y3[k + 1] + y3[k + 2])
    sum_value += to_sum
    k = k + 2
```

Valor da área de seção:

```python
print(f"{sum_value:.2f} m²")
```

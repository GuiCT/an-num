# Exerc√≠cio 2.

Definindo $X = (a + b)/2$ e realizando a substitui√ß√£o $ u = x - X $, com $du = dx$, transforma-se a integral para o intervalo dado por:

$$ \Rightarrow a - \frac{a + b}{2} = \frac{a - b}{2} = -L $$
e
$$ \Rightarrow b - \frac{a + b}{2} = \frac{b - a}{2} = L $$

dessa maneira, a integral √© transformada para esse novo intervalo:
$$ \int_a^b (x - X)^{n + 1}\ dx \Rightarrow \int_{-L}^L u^{n + 1}\ du $$

Como $n$ √© par, logo $n+1$ √© √≠mpar. Dessa maneira, $g(u) = u^{n + 1}$ √© uma fun√ß√£o √≠mpar, e a integral de uma fun√ß√£o √≠mpar em um intervalo sim√©trico em rela√ß√£o √† origem √© sempre zero:

$$ \int_{-L}^L u^{n + 1}\ du = 0 $$

O c√°lculo a partir de Newton-Cotes √© dada por uma express√£o do tipo

$$ Q = \sum_{k=0}^n w_k f(x_k) $$

Como definimos $X$ de forma que ele representa o ponto m√©dio do intervalo, existe a simetria:

$$ x_k - X = -(x_{n - k} - X) \Rightarrow x_k - X = X - x_{n - k} $$

Logo, a aplica√ß√£o $f(x_{n - k}) = (x_{n - k} - X)^{n + 1} = (-(x_k - X))^{n + 1}$ e a aplica√ß√£o $f(x_k) = (x_k - X)^{n + 1}$ est√£o relacionadas.

Como $f$ √© √≠mpar tal que $f(-x) = -f(x)$, obtemos a seguinte equival√™ncia:

$$ f(x_{n - k}) = -(x_k - X)^{n + 1} = -f(x_k) $$

Dada a f√≥rmula de Newton-Cotes:

$$ Q = \sum_{k = 0}^n w_k f(x_k) $$

Dado o termo central $x_{n/2}$ e sabendo da simetria dos pesos, √© poss√≠vel realizar a combina√ß√£o:

$$ Q = (w_0 f(x_0) + w_n f(x_n)) + (w_1 f(x_1) + w_{n-1} f(x_1)) + \dots + w_{n/2} f(x_{n/2}) $$

Pela simetria de pesos e simetria dos termos sim√©tricos:

$$ w_k f(x_k) + w_{n - k} f(x_n - k) = w_k f(x_k) + w_k (-f(x_k)) = w_k f(x_k) - w_k f(x_k) = 0 $$

E como $x_{n/2} = X$, logo $f(x_{n/2}) = (x_{n/2} - X)^{n + 1} = 0^{n + 1} = 0 $

Dessa maneira, confirmamos que a soma total da quadratura √© zero. Como $I=Q=0$, a f√≥rmula de Newton-Cotes √© exata para polin√¥mios de grau $n + 1$ quando $n$ √© par $\blacksquare$.

# Exerc√≠cio 3.

```python
x3 = np.arange(0, 22, 2, dtype=np.float64)
y3 = np.array([0, 1.8, 2, 4, 4, 6, 4, 3.6, 3.4, 2.8, 0])
h = x3[1] - x3[0]

# Utilizando f√≥rmula de Simpson Composta
sum_value = 0.0
k = 0
while k + 2 < len(y3):
    to_sum = (h / 3) * (y3[k] + 4 * y3[k + 1] + y3[k + 2])
    sum_value += to_sum
    k = k + 2
```

Valor da √°rea de se√ß√£o:

```python
print(f"{sum_value:.2f} m¬≤")
```
